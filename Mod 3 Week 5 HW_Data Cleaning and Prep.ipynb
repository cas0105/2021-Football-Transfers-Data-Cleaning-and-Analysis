```python
# create new dataframe that drops rows where a transfer to another club did not occur
transfers_updated = transfers[~transfers.team_dest.str.contains("Without Club")]
transfers_updated = transfers_updated[~transfers_updated.team_dest.str.contains("Retired")]
transfers_updated = transfers_updated[~transfers_updated.team_dest.str.contains("Unknown")]
transfers_updated = transfers_updated[~transfers_updated.team_dest.str.contains("Career break")]
transfers_updated = transfers_updated[~transfers_updated.team_dest.str.contains("Ban")]

transfers_updated
```

```python
print("The main data cleaning that needs to be done with this dataset involves the fee column.")
print("The values of the column do not conform to a standard.")
print("Most values have either m or th trailing a number to indicate if it is in millions or thousands.")
print("There are also missing values, text values that indicate a free transfer, and values of loan deals.")
print("\n")

# first, we will look at all unique values of the transfer_fee column
fees = transfers_updated.transfer_fee.unique()
print("Unique Values of the transfer_fee column:")
print(fees)
print("\n")

print("After looking at all unique values, the following should be done to clean the data:")
print("1. Remove any text values from the field.")
print("To do this, we will need to create a new column for indicating if a transfer is permanent or a loan.")
print("In addition, we will need to remove the text part from the loan deals.")
print("If a transfer is a free transfer, we need to change that value to 0.")
print("\n")
print("2. Transform all values to be the full value of transfer (ensure that no m or th trail numbers)")
print("This means we need to apply the appropriate multiplication to get the full values of transfers.")
```

```python
# create transfer type column
transfers_updated['transfer_type'] = np.where(transfers_updated['transfer_fee'].str.contains("loan", case=False, na=False), 'Loan', 'Permanent')
transfers_updated
```

```python
# to make sure the loan logic worked
transfers_updated.iloc[:,:][transfers_updated.transfer_type == 'Loan']
```

```python
# update values that should be 0 to zero
# draft
# loan transfer
# free transfer
# ?
# -
# NaN
# all of these need to be updated to 0

# first I will create a copy of the transfer_fee column to preserve the original data
transfers_updated['fee'] = transfers_updated['transfer_fee'].copy(deep=True)
transfers_updated['fee'] = transfers_updated['fee'].replace('draft', float(0.0))
transfers_updated['fee'] = transfers_updated['fee'].replace('loan transfer', float(0.0))
transfers_updated['fee'] = transfers_updated['fee'].replace('free transfer', float(0.0))
transfers_updated['fee'] = transfers_updated['fee'].replace('?', float(0.0))
transfers_updated['fee'] = transfers_updated['fee'].replace('-', float(0.0))
transfers_updated['fee'] = transfers_updated['fee'].replace(np.nan, float(0.0))

# next, we will remove the loan fee text
transfers_updated['fee'] = transfers_updated['fee'].replace({'Loan fee:': ''}, regex=True)

# we will also remove the Euro symbol
transfers_updated['fee'] = transfers_updated['fee'].replace({'€': ''}, regex=True)
```

```python
# now we will create a separate column for the units (millions or thousands)

# to define units of interest
unit = {"m":"millions","Th": "thousands"}

# create function to check if key is contained within string
def check_unit(x):
    for key in unit:
        if key.lower() in x.lower():
            return unit[key]
    return 'none'

# apply string to fee column
# if a fee has an m - column will indicate millions
# if a fee has Th - column will indicate thousands
transfers_updated.fee=transfers_updated.fee.astype(str)

transfers_updated['fee_unit'] = transfers_updated['fee'].map(lambda x: check_unit(x))
```

```python
# now we can update the fee column to remove the m and Th substrings from the column
# we will be left with the numerical value of the transfer
# then, we can use the unit column as a multiplyer for determining total value
transfers_updated['fee'] = transfers_updated['fee'].replace({'m': ''}, regex=True)
transfers_updated['fee'] = transfers_updated['fee'].replace({'Th.': ''}, regex=True)
transfers_updated.fee=transfers_updated.fee.astype(float)

def final_value(row):
    if row['fee_unit'] == 'millions':
        return row['fee']*1000000
    elif row['fee_unit'] == 'thousands':
        return row['fee']*1000
    else:
        return row['fee']
transfers_updated['transfer_fee_clean'] = transfers_updated.apply(lambda row: final_value(row), axis=1)
transfers_updated.transfer_fee_clean=transfers_updated.transfer_fee_clean.astype(float)
```

```python
# now lets check to see how many null values are in the updated dataset by column
transfers_updated.isna().sum()
```

```python
# one thing that pops out is that players without an orgin club have NaN for country and league orgin
# we can implement code to create a specified value for these
# first we will look for top recurring values of players not assigned to a club
transfers_updated['team_origin'].value_counts().nlargest(50, keep = 'all')
```

```python
# from this list, it looks like we need to focus on the following values:
# Without Club
# Unknown
# Career break
# Ban
# we will update the values of the country_origin and league_origin to say "None" if the team_origin is one of these values
# to define units of interest

def unknowns(row):
    if row['team_origin'] == 'Without Club':
        return 'Not Applicable'
    elif row['team_origin'] == 'Unknown':
        return 'Not Applicable'
    elif row['team_origin'] == 'Career break':
        return 'Not Applicable'
    elif row['team_origin'] == 'Ban':
        return 'Not Applicable'
    else:
        return row['country_origin']
transfers_updated['country_origin'] = transfers_updated.apply(lambda row: unknowns(row), axis=1)
transfers_updated['league_origin'] = transfers_updated.apply(lambda row: unknowns(row), axis=1)
transfers_updated
```

```python
# we can deal with the position nas by replacing the value with "Other"
transfers_updated['position'] = transfers_updated['position'].replace(np.nan, 'Other')
```

```python
# Part 4 - Handle Outliners
# Examine the data, filter out or transform outliners to meet the context of the data topic
# Provide informative descriptions and comments to the outliner data transformation opertions performaned in the context of the data topic

# Let's assume the analysis to be done is only on transfers where the value is greater than 0
# To do this, we will remove all outliers that have a clean transfer value of 0

col = transfers_updated['transfer_fee_clean']
free_trans = col[np.abs(col) == 0]
print("There are",len(free_trans),"records in the dataset in which the transfer was a free transfer or equal in value to 0.")
```

```python
# Now, we can create a new dataframe and removed all free transfers
transfers_updated2 = transfers_updated.copy(deep=True)
transfers_updated2 = transfers_updated2.iloc[:,:][transfers_updated2.transfer_fee_clean > 0]

# The other thing we can do to remove the impact of really high transfer fees is to transform the clean transfer fee value 
# we can create a new column that is the natural log of the clean transfer fee
transfers_updated2['ln(transfer_fee_clean)'] = np.log(transfers_updated2['transfer_fee_clean'])
transfers_updated2
```

```python
print("In the context of examining transfer fees of players being sold to another club:")
print("It makes sense to remove free transfers since those are often from player contracts expiring.")
print("It also makes sense to create a new column for the natural log of the transfer value to ensure large transfer fees do not spread the data too far apart.")
```
